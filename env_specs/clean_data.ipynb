```python
data = pd.read_csv('env_specs/eval19_data_wtruth.csv')
timings = pd.read_csv('env_specs/eval_timings_5April.csv')
timings.columns = ["analytic", "in", "done", "errors", "processed", "oo_loc", "oo_det", "oo_all", "avg_proc", "avg_nonproc"]
del timings['in'],timings['done'],timings['processed'],timings['oo_loc'],timings['oo_det'],timings['oo_all'],timings['avg_nonproc']
```

```python
data.describe()
```

```python
data_analytics = set([col[:-6] for col in data.columns if '_score' in col])
timing_analytics = set([analytic[:-16] for analytic in list(timings['analytic'])])
timings['analytic'] = timing_analytics
```

```python
timings.at[35, 'analytic'] = 'bingcamfinghdr'
timings.at[5, 'analytic'] = 'umd-weather'
```

```python
chosen = list(timing_analytics.intersection(data_analytics))
```

```python
# remove columns that are not necessary for preliminary tests
for column in data.columns:
    if '_status' in column:
        del data[column]
```

```python
todrop = []
for index, row in timings.iterrows():
    if row['analytic'] not in chosen:
        todrop.append(index)
timings.drop(timings.index[todrop], inplace=True)
timings.reset_index(drop=True, inplace=True)
```

### Fix timings

```python
import datetime
import time
for index, row in timings.iterrows():
    t = str(row['avg_proc'])
    if t != 'nan':
        x = time.strptime(t,'%H:%M:%S')
        y = datetime.timedelta(hours=x.tm_hour,minutes=x.tm_min,seconds=x.tm_sec).total_seconds()
    else:
        y = -1
    timings.at[index, 'avg_proc'] = y
```

```python
# sort by alphabetical 
timings = timings.sort_values('analytic', ascending=True).reset_index(drop=True)
```
